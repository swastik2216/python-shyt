```python
#A computer program is a list of "instructions" to be "executed" by a computer.
#In a programming language, these programming instructions are called statements.
#A program is made up of many statements that together perform a task.

print("Python is fun!")
print("Hello World!")
print("Have a good day.")
print("Learning Python is fun!")
```

```python
#Semicolons are optional in Python. You can write multiple statements on one line by separating them with ; but this is rarely used because it makes it hard to read:
x = 5; y = 10; print(x); print(y)
print("Hello"); print("How are you?"); print("Bye bye!")
```

```python
#Text in Python must be inside quotes. You can use either " double quotes or ' single quotes:
print("This will work!")
print('This will also work!')
```

```python
#A variable is created the moment you first assign a value to it.
x = 5
y = "John"
print(x)
print(y)
```

```python
#Variables do not need to be declared with any particular type, and can even change type after they have been set.
x = 4       # x is of type int
x = "Sally" # x is now of type str
print(x)
```

```python
#If you want to specify the data type of a variable, this can be done with casting.
x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0
print(x)
print(y)
print(z)
```

```python
#You can get the data type of a variable with the type() function.
x = 5
y = "John"
print(type(x))
print(type(y))
```

```python
#String variables can be declared either by using single or double quotes:
x = "John"
# is the same as
x = 'John'
```

```python
#Variable names are case-sensitive.
a = 4
A = "Sally"
#A will not overwrite and both variables will exist
print(a)
print(A)
```
