````python
While sitting in party, Tom came up with an idea of a quiz.
And the quiz is, Tom will spell out a number, and a person has to tell a number which is next to it. But
this number has to be a perfect square.write code
Here's a Python code that implements Tom's quiz:

```python
import math
def is_perfect_square(n):
    return int(math.sqrt(n)) ** 2 == n
def next_perfect_square(n):
    next_square = math.ceil(math.sqrt(n)) + 1
    return next_square ** 2
def quiz():
    number = int(input("Tom spells out a number: "))
    if is_perfect_square(number):
        print(f"The next perfect square after {number} is {next_perfect_square(number)}.")
    else:
        print(f"{number} is not a perfect square. Please try again.")
if __name__ == "__main__":    quiz()
```This code defines a function `is_perfect_square` to check if a number is a perfect square, and another function `next_perfect_square` to calculate the next perfect square after a given number. The `quiz` function prompts the user for a number, checks if it's a perfect square, and then provides the next perfect square if it is. If the input number is not a perfect square, it asks the user to try again.
You can run this code in a Python environment. When you execute the script, it will prompt you to enter a number. If you enter a perfect square (like 1, 4, 9, 16, etc.), it will display the next perfect square. If you enter a number that is not a perfect square, it will inform you and ask you to try again.
Here's the complete code for Tom's quiz:

```python

````

````python

def trap(height):
    left = 0
    right = len(height) - 1

    leftMax = 0
    rightMax = 0

    water = 0

    while left < right:

        if height[left] < height[right]:

            if height[left] >= leftMax:
                leftMax = height[left]
            else:
                water += leftMax - height[left]

            left += 1

        else:

            if height[right] >= rightMax:
                rightMax = height[right]
            else:
                water += rightMax - height[right]

            right -= 1

    return water


height = [4,2,0,3,2,5]
print(trap(height))
````
