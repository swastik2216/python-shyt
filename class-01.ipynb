```python
s = int(input("Enter the number: "))
if s%2 == 0:
    print("The number is even ")
else:
    print("The number is odd ")
```

```python
def isEven(n):

    # n|1 is greater than n, then even, else odd
    if (n | 1 > n):
        return True
    else:
        return False


# Driver code
if __name__ == "__main__":
    n = 100
    print("Even") if isEven(n) else print("Odd")
```
