```python
def migratoryBirds(arr):
    
    # ------------------------------------------------
    # STEP 1: Count frequencies using .get() 
    # (no Counter, no defaultdict - just a plain dict)
    # ------------------------------------------------
    counts = {}
    for bird in arr:
        counts[bird] = counts.get(bird, 0) + 1
    #  .get(bird, 0) means:
    #    if bird exists in dict -> return its current count
    #    if bird NOT in dict    -> return 0 (default)
    #  then we add 1 and store it back

    # ------------------------------------------------
    # STEP 2: Find the maximum count MANUALLY
    # (no max(), no sort() - just a loop)
    # ------------------------------------------------
    max_count = 0
    for bird in counts:
        if counts[bird] > max_count:
            max_count = counts[bird]

    # ------------------------------------------------
    # STEP 3: Find the winner MANUALLY
    # If tie -> smallest ID wins
    # We start with a large number and keep replacing
    # ------------------------------------------------
    result = 999999999          # a very large placeholder
    for bird in counts:
        if counts[bird] == max_count:
            if bird < result:   # keep the smallest ID
                result = bird

    return result


# ---- Driver code ----
arr = [1, 2, 3, 4, 5, 4, 3, 2, 1, 3, 4]
print("Answer:", migratoryBirds(arr))   # Expected: 3
```

```python
def reverse(arr, l, r):
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l += 1
        r -= 1

def rotate(arr, k):
    n = len(arr)
    k = k % n  

    reverse(arr, 0, n - 1)   
    reverse(arr, 0, k - 1)   
    reverse(arr, k, n - 1)   

arr = [1, 2, 7, 16, 12]
k = 3
rotate(arr, k)
print(arr)  
```

```python
#integer problme 
def find_unique(arr):
    result = 0
    for num in arr:
        result ^= num  
    return result


arr = [4, 3, 2, 4, 1, 3, 2]
print(find_unique(arr))  
```

```python
def sock_pairs(socks):
    freq = {}
    for color in socks:
        freq[color] = freq.get(color, 0) + 1

    pairs = 0
    for count in freq.values():
        pairs += count // 2

    return pairs

socks = [10, 20, 20, 10, 10, 30, 50, 10, 20]
print(sock_pairs(socks)) 
```
