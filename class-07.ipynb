```python
from math import gcd
from functools import reduce

def lcm(a, b):
    return a * b // gcd(a, b)

def getTotalX(a, b):
    l = reduce(lcm, a)
    g = reduce(gcd, b)

    count = 0
    for i in range(l, g + 1, l):
        if g % i == 0:
            count += 1

    return count
print(getTotalX([2, 4, 8], [16, 32, 96])) 
```

```python
#count the no of vowels and consonants in a string
def count_vowels_consonants(s):
    vowels = 'aeiouAEIOU'
    vowel_count = 0
    consonant_count = 0

    for char in s:
        if char.isalpha():  # Check if the character is an alphabet
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    return vowel_count, consonant_count
# Example usage
string = "Hello World!"
vowels, consonants = count_vowels_consonants(string)
print(f"Vowels: {vowels}, Consonants: {consonants}")



```

```python
#print the number of prime numbers from 2 to 30
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
def count_primes(start, end):
    prime_count = 0
    for num in range(start, end + 1):
        if is_prime(num):
            prime_count += 1
    return prime_count
print(count_primes(2, 30))
```
