```python
from math import gcd
from functools import reduce

def lcm(a, b):
    return a * b // gcd(a, b)

def getTotalX(a, b):
    l = reduce(lcm, a)
    g = reduce(gcd, b)

    count = 0
    for i in range(l, g + 1, l):
        if g % i == 0:
            count += 1

    return count
print(getTotalX([2, 4, 8], [16, 32, 96])) 
```
