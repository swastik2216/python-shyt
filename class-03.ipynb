```python
def dec_to_hex_manual(n):
    if n == 0:
        return "0"
    
    hex_chars = "0123456789ABCDEF"
    result = ""
    
    while n > 0:
        remainder = n % 16
        result = hex_chars[remainder] + result
        n = n // 16
        
    return result

print(dec_to_hex_manual(254))  
```

```python
#an integer d is a divisor of an integer n if n % d = 0.given an integer, 
#for each digit that makes up the integer deteminewhether 
#if it is a divisor .
def count_divisors(n):
    count = 0
    for digit in str(n):
        if digit != '0':  
            if n % int(digit) == 0:
                count += 1
    return count
print(count_divisors(123))  
```

```python
#sweet seventeen 
def decimal_to_base17(decimal_num):
    if decimal_num == 0:
        return "0"

    base17_digits = "0123456789ABCDEFG"
    result = ""

    while decimal_num > 0:
        remainder = decimal_num % 17
        result = base17_digits[remainder] + result
        decimal_num //= 17

    return result
print(decimal_to_base17(27))  # Output: '1A'

print(int("1A", 17))  
```

```python
#iven two non negative integers n1 and n2, where n1
#suppose n1 =11 and n2=15
#there is the number 11, which haas repeated digits but 12,13,14,15 do not have repeated digits.
#so the output it 4
def count_unique_digit_numbers(n1, n2):
    count = 0

    for num in range(n1, n2 + 1):
        num_str = str(num)
        if len(num_str) == len(set(num_str)):
            count += 1

    return count
print(count_unique_digit_numbers(101, 200))  
```

```python
sereis:0,0,2,1,4,2,6,3,8,4,10,5,12,6,14,7,16,8,18,9
def generate_series(n):
    series = []
    for i in range(n):
        if i % 2 == 0:
            series.append(i)
        else:
            series.append(i // 2)
    return series
print(generate_series(20))
```

```python
def find_nth_term(n):
    if n % 2 == 0:
        return n
    else:
        return n // 2
print(find_nth_term(17))  
```
